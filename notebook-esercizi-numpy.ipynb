### Esercizio 2

Generare un array Numpy che contenga le posizioni dei primi 100 numeri primi all'interno dei numeri naturali.

```python
def isPrime(num):
    res = True
    
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            res = False
            break
            
    return res
```

```python
isPrime_v = np.vectorize(isPrime)
natural_n = np.arange(2, 1000)
```

```python
(natural_n[isPrime_v(natural_n)])[:100] + 1
```

### Esercizio 15

Caricare i dati meteo dai file CSV e calcolare usando numpy:
- n° di giorni totali senza neve al suolo nelle due stazioni.
- n° di giorni nell'anno 2020 in cui la media delle temperature medie delle due stazioni è minore o uguale a zero.
- n° di giorni nell'anno 2019 in cui c'è neve al suolo e la temperatura media è maggiore di 0°C per entrambe le stazioni.

```python
stazione1 = {
    "neve": {},
    "temperature": {}
}

stazione2 = {
    "neve": {},
    "temperature": {}
}
```

```python
def readCSV(path, n_line_to_skip, dtypes):
    res = {}
    
    with open(path) as csv_file:
        for _ in range(n_line_to_skip):
            csv_file.readline()
            
        csv_reader = csv.reader(csv_file, delimiter=",")
        
        for curr_line, row in enumerate(csv_reader):
            row = np.array([s.strip() for s in row])

            if curr_line == 0:
                res = {header: np.zeros(0) for header in row}
            else:
                if '' in row:
                    row[row == ''] = '-9999'
                
                for i, field in enumerate(row):
                    res[list(res.keys())[i]] = np.append(res[list(res.keys())[i]], dtypes[i](field))
    
    res["data"] = res["data"][:-1]
                    
    return res
```

```python
stazione1["temperature"] = readCSV('./stazioni_metereologiche/richiesta_dati_gg_80749_parte1.csv', 9, [str, str, float, str, float, str, str, float, str])
stazione1["neve"] = readCSV('./stazioni_metereologiche/richiesta_dati_gg_80749_parte2.csv', 9, [str, int, int])
stazione2["temperature"] = readCSV('./stazioni_metereologiche/richiesta_dati_gg_80749_parte3.csv', 9, [str, str, float, str, float, str, str, float, str])
stazione2["neve"] = readCSV('./stazioni_metereologiche/richiesta_dati_gg_80749_parte4.csv', 9, [str, int, int])
```

```python
np.count_nonzero(stazione1["neve"]["neve al suolo (cm)"] == 0)
```

```python
np.count_nonzero(stazione2["neve"]["neve al suolo (cm)"] == 0)
```

```python
med = (
    stazione1["temperature"]["Tmed"][np.char.find(stazione1["temperature"]["data"], "2020") != -1] +
    stazione2["temperature"]["Tmed"][np.char.find(stazione2["temperature"]["data"], "2020") != -1]
) / 2

np.count_nonzero(np.logical_and(med <= 0, med > -9999))
```

```python
filtro_2019 = np.intersect1d(
    stazione1["neve"]["data"][np.char.find(stazione1["neve"]["data"], "2019") != -1],
    stazione1["temperature"]["data"][np.char.find(stazione1["temperature"]["data"], "2019") != -1]
)

filtro_tmp = np.where(np.in1d(stazione1["temperature"]["data"], filtro_2019))
filtro_nev = np.where(np.in1d(stazione1["neve"]["data"], filtro_2019))

np.count_nonzero(
    np.logical_and(
        stazione1["temperature"]["Tmed"][filtro_tmp] > 0,
        stazione1["neve"]["neve al suolo (cm)"][filtro_nev] > 0
    )
)
```

```python
filtro_2019 = np.intersect1d(
    stazione2["neve"]["data"][np.char.find(stazione2["neve"]["data"], "2019") != -1],
    stazione2["temperature"]["data"][np.char.find(stazione2["temperature"]["data"], "2019") != -1]
)

filtro_tmp = np.where(np.in1d(stazione2["temperature"]["data"], filtro_2019))
filtro_nev = np.where(np.in1d(stazione2["neve"]["data"], filtro_2019))

np.count_nonzero(
    np.logical_and(
        stazione2["temperature"]["Tmed"][filtro_tmp] > 0,
        stazione2["neve"]["neve al suolo (cm)"][filtro_nev] > 0
    )
)
```

### Esercizio 16

Nel file csv allegato sono riportate le variazioni percentuali giornaliere dei prezzi di alcune azioni, per esempio 5% corrisponde a 0.05.
La prima colonna è la data, quelle successive sono le  variazioni percentuali giornaliere dei prezzi dell'azione corrispondente. I nomi colonna sono del tipo: TICKER% dove TICKER è il ticker dell'azione. Il ticker può essere associato alla azienda tramite motori di ricerca  come questo: https://it.investing.com

Caricare i dati in opportuni array numpy e:
- calcolare il numero di giorni in cui Alphabet va meglio di Tesla
- calcolare i massimi locali nella percentuale giornaliera di Meta Platforms
- calcolare se per Microsoft Corporation i giorni positivi sono più di quelli negativi
- calcolare quanti sono i giorni in cui Microsoft Corporation è positiva e Apple è negativa
- stampare in quali date la somma delle variazioni di prezzo di Tesla e Apple è maggiore di 0
- trovare se ci sono periodi di 3 giorni consecutivi in cui Amazon.com è negativa e in caso affermativo in che date iniziano
- fare un grafico che rappresenti l'andamento delle  variazioni percentuali giornaliere  di Amazon.com
- fare un istogramma che rappresenti le  variazioni percentuali giornaliere  di Tesla.

```python
dati = {}

with open("./analisi_finanziaria/dati_azioni.csv") as csv_file:    
    csv_reader = csv.reader(csv_file, delimiter=",")
    dtypes = [str, float, float, float, float, float, float]
    
    for curr_line, row in enumerate(csv_reader):
        if curr_line == 0:
            dati = {header: np.zeros(0) for header in row}
        else:
            for i, field in enumerate(row):
                dati[list(dati.keys())[i]] = np.append(dati[list(dati.keys())[i]], dtypes[i](field))
```

```python
np.count_nonzero(dati["GOOG%"] > dati["TSLA%"])
```

```python
np.count_nonzero(
    np.logical_and(
        dati["FB%"] > dati["AMZN%"],
        dati["FB%"] > dati["TSLA%"]
    )
)
```

```python
pos = np.count_nonzero(dati['MSFT%'] > 0)
neg = np.count_nonzero(dati['MSFT%'] < 0)

if pos > neg:
    print(f"Più giorni positivi! {pos}")
elif pos < neg:
    print(f"Più giorni negativi! {neg}")
else:
    print(f"Stesso numero di giorni! pos: {pos} neg: {neg}")
```

```python
np.count_nonzero(
    np.logical_and(
        dati['MSFT%'] > 0,
        dati['AAPL%'] < 0
    )
)
```

```python
dati["Date"][(dati['TSLA%'] + dati["AAPL%"]) > 0]
```

```python
amzn1 = np.roll(dati["AMZN%"], -1)
amzn2 = np.roll(dati["AMZN%"], -2)

filtro = np.logical_and(
    np.logical_and(
        amzn1 < 0,
        amzn2 < 0
    ), 
    dati["AMZN%"] < 0
)

dati["Date"][filtro]
```

```python
str_to_date_v = np.vectorize(dt.datetime.strptime)

x = str_to_date_v(dati["Date"], '%Y-%m-%d')
```

```python
y = dati["AMZN%"]

fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(x, y, "y")
ax.set_title("Variazioni Percentuali Giornaliere Amazon.com")
ax.set_xlabel("Date")
ax.set_ylabel("Variazioni percentuali")
plt.show()
```

```python
#fare un istogramma che rappresenti le variazioni percentuali giornaliere di Tesla.
y = dati["TSLA%"]

fig, ax = plt.subplots(figsize=(20, 10))
ax.hist(y, bins=y.size, color="red")
ax.set_title("Variazioni Percentuali Giornaliere Tesla")
ax.set_xlabel("Variazioni percentuali")
ax.set_ylabel("Frequenza")
plt.show()
```
