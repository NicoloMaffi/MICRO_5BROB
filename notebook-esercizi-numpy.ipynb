### Esercizio 2

Generare un array Numpy che contenga le posizioni dei primi 100 numeri primi all'interno dei numeri naturali.

```python
def isPrime(num):
    res = True
    
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            res = False
            break
            
    return res
```

```python
isPrime_v = np.vectorize(isPrime)
natural_n = np.arange(2, 1000)
```

```python
(natural_n[isPrime_v(natural_n)])[:100] + 1
```

### Esercizio 14

Sul territorio comunale sono installate tre centraline per la misurazione del livello di polveri sottili PM10 nell'aria. Talvolta le centraline vanno in uno stato di errore per cui la misurazione delle PM10 fallisce ed il dato corrispondente riporta il valore 9999.9 all'interno dell'array delle misurazioni. Inoltre, per motivi sconosciuti, talvolta le centraline riportano il valore -1. Statisticamente si è visto che il primo errore ha una frequenza dello 0.2% mentre il secondo del 0.1%

- Creare una matrice avente 3 colonne e le righe contenente i dati di tutte le centraline registrati nelle 24 ore (si registra il dato ogni minuto)
- Eliminare tutte le righe della matrice in cui vi sia almeno una misura errata o una misura di origine sconosciuta

```python
n_mis = 24 * 60

mat = np.random.uniform(0, 100, size=(n_mis, 3))
mat
```

```python
mat[np.random.choice(range(n_mis), round(n_mis * 0.2/100), replace=False), np.random.randint(0, 3)] = 9999.99
mat[np.random.choice(range(n_mis), round(n_mis *0.1/100), replace=False), np.random.randint(0, 3)] = -1
mat
```

```python
np.count_nonzero(np.logical_or(mat == 9999.99, mat == -1))
```

```python
res = np.copy(mat[np.logical_not(np.any(np.logical_or(mat == 9999.99, mat == -1), axis=1))])
```

```python
np.count_nonzero(np.logical_or(res == 9999.99, res == -1))
```

### Esercizio 15

Caricare i dati meteo dai file CSV e calcolare usando numpy:
- n° di giorni totali senza neve al suolo nelle due stazioni.
- n° di giorni nell'anno 2020 in cui la media delle temperature medie delle due stazioni è minore o uguale a zero.
- n° di giorni nell'anno 2019 in cui c'è neve al suolo e la temperatura media è maggiore di 0°C per entrambe le stazioni.

```python
stazione1 = {
    "neve": {},
    "temperature": {}
}

stazione2 = {
    "neve": {},
    "temperature": {}
}
```

```python
def readCSV(path, n_line_to_skip, dtypes):
    res = {}
    
    with open(path) as csv_file:
        #Vengono rimosse le prime righe che non fanno parte dei dati
        for _ in range(n_line_to_skip):
            csv_file.readline()
            
        csv_reader = csv.reader(csv_file, delimiter=",")
        
        for curr_line, row in enumerate(csv_reader):
            #Rimozioni caratteri di escape e blank a dx e sx in ogni stringa del record
            row = [s.strip() for s in row]

            if curr_line == 0:
                #Prima linea -> nomi delle colonne
                res = {header: np.zeros(0) for header in row}
            elif '' in row:
                #Record contenenti errori
                continue
            else:
                for i, field in enumerate(row):
                    """f_type = dtypes[i] #Data type del campo letto (field)
                    
                    #Se il data type è numerico è field è vuoto allora viene scritto 0
                    if field == '' and (f_type is int or f_type is float):
                        field = '0'"""
                    
                    #Aggiunta e conversione del campo nell'array corrispondente
                    res[list(res.keys())[i]] = np.append(res[list(res.keys())[i]], dtypes[i](field))
    
    #Rimozione ultima riga che non fa parte dei dati
    res["data"] = res["data"][:-1]
                    
    return res
```

```python
#Lettura dati
stazione1["temperature"] = readCSV('./richiesta_dati_gg_80749_parte1.csv', 9, [str, str, float, str, float, str, str, float, str])
stazione1["neve"] = readCSV('./richiesta_dati_gg_80749_parte2.csv', 9, [str, int, int])
stazione2["temperature"] = readCSV('./richiesta_dati_gg_80749_parte3.csv', 9, [str, str, float, str, float, str, str, float, str])
stazione2["neve"] = readCSV('./richiesta_dati_gg_80749_parte4.csv', 9, [str, int, int])
```

```python
#Task 3 - Stazione 1

#Dato che stazioneX["neve"] e stazioneX["temperature"] non hanno lo stesso numero di record (stazioneX["neve"] è + piccolo)
#allora si aggiunge in coda all'array di bool "stazione1["neve"]["neve al suolo (cm)"] > 0" un numero di elementi
#pari alla differenza dei record, dove ogni elemento è FALSE in modo tale che questi ultimi non vengano inclusi nel risultato

diff = np.abs(stazione1["neve"]["neve al suolo (cm)"].size - stazione1["temperature"]["data"].size)

np.count_nonzero(
    np.logical_and(
        np.logical_and(
            np.char.find(stazione1["temperature"]["data"], "2019") != -1,
            np.append(stazione1["neve"]["neve al suolo (cm)"] > 0, np.zeros(diff, dtype=bool))
        ),
        stazione1["temperature"]["Tmed"] > 0
    )
)
```
